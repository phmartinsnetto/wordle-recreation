<h1>Data Cleaning: English words list from 12dicts</h1>

<h2>Loading word list</h2>

```python
# Reading words file and attributing it to a variable
with open('.path to your file\2of12inf.txt', 'r') as file:
    complete_wordlist = file.read()
    
    
complete_wordlist
```

```python
# Our wordlist is a long string, the words separated by \n
# we have to evoke the String.split(sep = '\n') method so we can separate the words in different elements of a list,
# cosidering \n as a separator.

complete_wordlist = complete_wordlist.split(sep = '\n')
complete_wordlist
```

```python
# importing pandas and converting list to a dataframe.

import pandas as pd

complete_wordlist_df = pd.DataFrame(complete_wordlist)
complete_wordlist_df.columns = ['Word']
complete_wordlist_df.head()
```

```python
# Selecting only words with 5 letters
# .apply() method applies a function to a dataframe axis, in this case it'll be the .len()

selection = complete_wordlist_df['Word'].apply(len) == 5
selection
```

```python
# Creating a new dataframe corresponding to ou 5 letters selection

filtered_wordlist_df = complete_wordlist_df[selection]
filtered_wordlist_df
```

```python
# Resetting the index

filtered_wordlist_df.reset_index(drop = True, inplace = True)
filtered_wordlist_df
```

```python
# Converting Words column to Series

filtered_wordlist = filtered_wordlist_df['Word']
filtered_wordlist.head()
```

```python
# Exporting our filtered series to a .JSON file

filtered_wordlist.to_json('.directory where you want to save the file\filtered_wordlist.json')
```
